```python
import math
x=0
y=0
eta=0.00001
eps=0.1
delx=1
dely=1
e=2.178
max_iter=1000
iter=0
def deriv(x,y):
  xderive=6*x
  yderive=-5*pow(e,-y)
  return xderive,yderive
```

```python
while max(abs(delx),abs(dely))>eps and iter<max_iter:
  prevx=x
  prevy=y
  delx,dely=deriv(prevx,prevy)
  delx=-eta*delx
  dely=-eta*dely
  x=x+delx
  y=y+dely
  iter=iter+1
print("Value of x = ",round(x),"\nValue of y = ",round(y));
obj=3*x*x+5*pow(e,-y)+10
print("value of objective function is ",round(obj))
```
